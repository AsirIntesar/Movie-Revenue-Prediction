```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data= pd.read_csv('cost_revenue_clean.csv')
data
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.6)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

```python
regression.coef_  #Slope Coefficient
```

```python
regression.intercept_  #Intercept
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.6)

#Regression.predict(X) will plot a line predicting each of the point of X instead worldwide gross
plt.plot(X, regression.predict(X), color='red', linewidth=4) 
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
regression.score(X,y) #percentage of revenue is here 55% which is pretty good and we can say final result is in profit
```
